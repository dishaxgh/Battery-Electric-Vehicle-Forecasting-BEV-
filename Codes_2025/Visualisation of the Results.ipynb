```python
# Importing Trial 3 files
T3_lr_followers = pd.read_excel(r"C:\Users\Disha\TU Braunschweig\Thesis\Codes_2025\Trial 3\Trial3_LR_Followers_Results.xlsx")
T3_lr_leaders = pd.read_excel(r"C:\Users\Disha\TU Braunschweig\Thesis\Codes_2025\Trial 3\Trial3_LR_Leaders_Results.xlsx")

T3_svr_followers = pd.read_excel(r"C:\Users\Disha\TU Braunschweig\Thesis\Codes_2025\Trial 3\Trial3_SVR_Followers_Results.xlsx")
T3_svr_leaders = pd.read_excel(r"C:\Users\Disha\TU Braunschweig\Thesis\Codes_2025\Trial 3\Trial3_SVR_Leaders_Results.xlsx")

T3_ARIMAX_followers = pd.read_excel(r"C:\Users\Disha\TU Braunschweig\Thesis\Codes_2025\Trial 3\Trial3_ARIMAX_Followers_Results.xlsx")
T3_ARIMAX_leaders = pd.read_excel(r"C:\Users\Disha\TU Braunschweig\Thesis\Codes_2025\Trial 3\Trial3_ARIMAX_Leaders_Results.xlsx")

T3_rf_followers = pd.read_excel(r"C:\Users\Disha\TU Braunschweig\Thesis\Codes_2025\Trial 3\Trial3_RF_Followers_Results.xlsx")
T3_rf_leaders = pd.read_excel(r"C:\Users\Disha\TU Braunschweig\Thesis\Codes_2025\Trial 3\Trial3_RF_Leaders_Results.xlsx")

T3_xgb_followers = pd.read_excel(r"C:\Users\Disha\TU Braunschweig\Thesis\Codes_2025\Trial 3\Trial3_XGB_Followers_Results.xlsx")
T3_xgb_leaders = pd.read_excel(r"C:\Users\Disha\TU Braunschweig\Thesis\Codes_2025\Trial 3\Trial3_XGB_Leaders_Results.xlsx")

T3_lstm_followers = pd.read_excel(r"C:\Users\Disha\TU Braunschweig\Thesis\Codes_2025\Trial 3\Trial3_LSTM_Followers_Results.xlsx")
T3_lstm_leaders = pd.read_excel(r"C:\Users\Disha\TU Braunschweig\Thesis\Codes_2025\Trial 3\Trial3_LSTM_Leaders_Results.xlsx")

T3_proposed_ElasticNet_followers = pd.read_excel(r"C:\Users\Disha\TU Braunschweig\Thesis\Codes_2025\Trial 3\Trial3_Proposed (ElasticNet)_Followers_Results.xlsx")
T3_proposed_ElasticNet_leaders = pd.read_excel(r"C:\Users\Disha\TU Braunschweig\Thesis\Codes_2025\Trial 3\Trial3_Proposed (ElasticNet)_Leaders_Results.xlsx")

T3_proposed_GradientBoost_followers = pd.read_excel(r"C:\Users\Disha\TU Braunschweig\Thesis\Codes_2025\Trial 3\Trial3_Proposed (Gradient Boost)_Followers_Results.xlsx")
T3_proposed_GradientBoost_leaders = pd.read_excel(r"C:\Users\Disha\TU Braunschweig\Thesis\Codes_2025\Trial 3\Trial3_Proposed (Gradient Boost)_Leaders_Results.xlsx")

T3_proposed_Weighted_followers = pd.read_excel(r"C:\Users\Disha\TU Braunschweig\Thesis\Codes_2025\Trial 3\Trial3_Proposed (Weighted)_Followers_Results.xlsx")
T3_proposed_Weighted_leaders = pd.read_excel(r"C:\Users\Disha\TU Braunschweig\Thesis\Codes_2025\Trial 3\Trial3_Proposed (Weighted)_Leaders_Results.xlsx")
```

```python
# Helper function to add metadata columns
def add_metadata(df, trial, model, cluster_type):
    df = df.copy()
    df["Trial"] = trial
    df["Model"] = model
    df["ClusterType"] = cluster_type
    return df

# Combine all datasets
all_results = pd.concat([

    # TRIAL 3 
    add_metadata(T3_lr_followers, "Trial 3", "Linear Regression", "Followers"),
    add_metadata(T3_lr_leaders, "Trial 3", "Linear Regression", "Leaders"),
    add_metadata(T3_svr_followers, "Trial 3", "SVR", "Followers"),
    add_metadata(T3_svr_leaders, "Trial 3", "SVR", "Leaders"),
    add_metadata(T3_ARIMAX_followers, "Trial 3", "ARIMAX", "Followers"),
    add_metadata(T3_ARIMAX_leaders, "Trial 3", "ARIMAX", "Leaders"),
    add_metadata(T3_rf_followers, "Trial 3", "RF", "Followers"),
    add_metadata(T3_rf_leaders, "Trial 3", "RF", "Leaders"),
    add_metadata(T3_xgb_followers, "Trial 3", "XGB", "Followers"),
    add_metadata(T3_xgb_leaders, "Trial 3", "XGB", "Leaders"),
    add_metadata(T3_lstm_followers, "Trial 3", "LSTM", "Followers"),
    add_metadata(T3_lstm_leaders, "Trial 3", "LSTM", "Leaders"),
    add_metadata(T3_proposed_ElasticNet_followers, "Trial 3", "Proposed (ElasticNet)", "Followers"),
    add_metadata(T3_proposed_ElasticNet_leaders, "Trial 3", "Proposed (ElasticNet)", "Leaders"),
    add_metadata(T3_proposed_GradientBoost_followers, "Trial 3", "Proposed (GradientBoost)", "Followers"),
    add_metadata(T3_proposed_GradientBoost_leaders, "Trial 3", "Proposed (GradientBoost)", "Leaders"),
    add_metadata(T3_proposed_Weighted_followers, "Trial 3", "Proposed (Weighted)", "Followers"),
    add_metadata(T3_proposed_Weighted_leaders, "Trial 3", "Proposed (Weighted)", "Leaders")
], ignore_index=True)

# Verification
print(all_results.shape)
print(all_results.columns)
all_results.head()
```

## **Note:**

Portugal will be present in leaders and followers both because - 

Portugal is in **"LEADERS" for "Trial 1 and Trial 2"**

Portugal is in **"FOLLOWERS" FOR "Trial 3"**

```python
# Model we want to plot
model_name = "Linear Regression"

# Filter data for the chosen model
df_model = all_results[all_results["Model"] == model_name]

# Compute mean absolute error per Country × Trial × Cluster
mean_error = (
    df_model.groupby(["Country", "Trial", "ClusterType"])["Error_Abs"]
    .mean()
    .reset_index()
)

# Separate leaders and followers for clear plotting
leaders = mean_error[mean_error["ClusterType"] == "Leaders"]
followers = mean_error[mean_error["ClusterType"] == "Followers"]

# Sort countries alphabetically or by average error if you prefer
leaders = leaders.sort_values("Country")
followers = followers.sort_values("Country")

# Visualise
sns.set(style="whitegrid", font_scale=1.1)
fig, axes = plt.subplots(2, 1, figsize=(15, 10), sharey=False)

# Leaders Plot
sns.barplot(
    data=leaders,
    x="Country",
    y="Error_Abs",
    hue="Trial",
    palette="crest",
    ax=axes[0]
)
axes[0].set_title(f"{model_name}: Mean Absolute Error by Trial (LEADERS)", fontsize=14)
axes[0].set_ylabel("Mean Error (BEV %)")
axes[0].set_xlabel("")
axes[0].tick_params(axis='x', rotation=45)

# Followers Plot 
sns.barplot(
    data=followers,
    x="Country",
    y="Error_Abs",
    hue="Trial",
    palette="rocket",
    ax=axes[1]
)
axes[1].set_title(f"{model_name}: Mean Absolute Error by Trial (FOLLOWERS)", fontsize=14)
axes[1].set_ylabel("Mean Absolute Error (BEV %)")
axes[1].set_xlabel("Country")
axes[1].tick_params(axis='x', rotation=45)

for ax in axes:
    ax.legend(title="Trial", loc="upper right")
    ax.set_ylim(0, max(mean_error["Error_Abs"]) * 1.1)  
plt.tight_layout()
plt.savefig("Linear Regression_2025 (Leaders and Followers)_MEAN.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
