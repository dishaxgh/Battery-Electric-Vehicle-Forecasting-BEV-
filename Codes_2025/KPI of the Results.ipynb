```python
# Helper function to add metadata columns
def add_metadata(df, trial, model, cluster_type):
    df = df.copy()
    df["Trial"] = trial
    df["Model"] = model
    df["ClusterType"] = cluster_type
    return df

# Combine all datasets
all_results = pd.concat([

    # TRIAL 3 
    add_metadata(T3_AutoML_followers, "Trial 3", "AutoGluon", "Followers"),
    add_metadata(T3_AutoML_leaders, "Trial 3", "AutoGluon", "Leaders"),
    add_metadata(T3_lr_followers, "Trial 3", "Linear Regression", "Followers"),
    add_metadata(T3_lr_leaders, "Trial 3", "Linear Regression", "Leaders"),
    add_metadata(T3_svr_followers, "Trial 3", "SVR", "Followers"),
    add_metadata(T3_svr_leaders, "Trial 3", "SVR", "Leaders"),
    add_metadata(T3_ARIMAX_followers, "Trial 3", "ARIMAX", "Followers"),
    add_metadata(T3_ARIMAX_leaders, "Trial 3", "ARIMAX", "Leaders"),
    add_metadata(T3_rf_followers, "Trial 3", "RF", "Followers"),
    add_metadata(T3_rf_leaders, "Trial 3", "RF", "Leaders"),
    add_metadata(T3_xgb_followers, "Trial 3", "XGB", "Followers"),
    add_metadata(T3_xgb_leaders, "Trial 3", "XGB", "Leaders"),
    add_metadata(T3_lstm_followers, "Trial 3", "LSTM", "Followers"),
    add_metadata(T3_lstm_leaders, "Trial 3", "LSTM", "Leaders"),
    add_metadata(T3_proposed_ElasticNet_followers, "Trial 3", "Proposed (ElasticNet)", "Followers"),
    add_metadata(T3_proposed_ElasticNet_leaders, "Trial 3", "Proposed (ElasticNet)", "Leaders"),
    add_metadata(T3_proposed_GradientBoost_followers, "Trial 3", "Proposed (GradientBoost)", "Followers"),
    add_metadata(T3_proposed_GradientBoost_leaders, "Trial 3", "Proposed (GradientBoost)", "Leaders"),
    add_metadata(T3_proposed_Weighted_followers, "Trial 3", "Proposed (Weighted)", "Followers"),
    add_metadata(T3_proposed_Weighted_leaders, "Trial 3", "Proposed (Weighted)", "Leaders")
], ignore_index=True)

# Verification
print(all_results.shape)
print(all_results.columns)
all_results.head()
```

```python
# KPI - Median Absolute Error
# Reason to use this - to find the "middle" error, ignoring extreme outliers
# We group by Trial, Cluster, and Model to find the typical error performance
medae_results = all_results.groupby(['Trial', 'ClusterType', 'Model'])['Error_Abs'].median().reset_index()
medae_results.columns = ['Trial', 'Cluster', 'Model', 'MedAE (%)']

pivot_medae = medae_results.pivot_table(
    index=['Cluster', 'Model'], 
    columns='Trial', 
    values='MedAE (%)'
)

# Sort in ascending order of the Results 
pivot_medae = pivot_medae.sort_values(by=['Cluster', 'Trial 3'], ascending=True)

# Print Results
print("Trial 3 Case Study - Tested on year 2025")
print(pivot_medae)
```

```python
# KPI - Mean Absolute Error
# We group by Trial, Cluster, and Model to find the typical error performance
mae_results = all_results.groupby(['Trial', 'ClusterType', 'Model'])['Error_Abs'].mean().reset_index()
mae_results.columns = ['Trial', 'Cluster', 'Model', 'MAE (%)']

pivot_mae = mae_results.pivot_table(
    index=['Cluster', 'Model'], 
    columns='Trial', 
    values='MAE (%)'
)

# Sort in ascending order of the Results 
pivot_mae = pivot_mae.sort_values(by=['Cluster', 'Trial 3'], ascending=True)

# Print Results
print("Trial 3 Case Study - Tested on year 2025")
print(pivot_mae)
```
