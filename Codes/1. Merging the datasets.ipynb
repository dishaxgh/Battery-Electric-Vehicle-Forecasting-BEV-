```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
import os
```

**Step 1: Merging all country files in the target folder: AF Market share of total registrations (M1)**

```python
# Path to all country-level Excel files
target_files_marketShare = glob.glob(r"C:\Users\Disha\TU Braunschweig\Thesis\Data\Target\AF Market share of total registrations (M1)\*.csv")

# To make sure the files are present
print("Files found:", len(target_files_marketShare))
print(target_files_marketShare[:5]) 
```

```python
target_list = []

for file in target_files_marketShare:
    
    # Extract country name from filename 
    country = os.path.basename(file).split(".")[0]
    
    # Read the CSV/Excel file
    df = pd.read_csv(file)
    # If Excel files, use: pd.read_excel(file)
    
    # Rename the first column to 'Year' (previously 'Category')
    df = df.rename(columns={df.columns[0]: "Year"})
    
    # Keep only required columns
    df = df[["Year", "BEV"]].copy()
    
    # Clean up 'Country' values (remove "(Target)" if it exists)
    df["Country"] = country.replace("(Target)", "").strip()
    
    # Ensure Year is numeric
    df = df[pd.to_numeric(df["Year"], errors="coerce").notnull()]
    df["Year"] = df["Year"].astype(int)
    
    # Rearranging the columns
    df = df[["Country", "Year", "BEV"]]
    
    # Append to the list
    target_list.append(df)

# Combine all countries into one DataFrame
targets_marketShare = pd.concat(target_list, ignore_index=True)

print("Target data combined. Shape:", targets_marketShare.shape)
display(targets_marketShare.head())
```

```python
targets_marketShare['Country'].unique()
```

```python
# Keep only those countries those are part of EU27
countries_to_remove = [
    'European Union (EU27)', 'Iceland', 'Liechtenstein', 'Norway',
    'Switzerland', 'Turkey', 'United Kingdom'
]

targets_marketShare = targets_marketShare[~targets_marketShare['Country'].isin(countries_to_remove)]

targets_marketShare['Country'].unique()
```

**Step 6: Recharging Points**

```python
# Path to all country-level Excel files
recharge_files_2019 = glob.glob(r"C:\Users\Disha\TU Braunschweig\Thesis\Data\Parameters\Recharging points (2012-2019)\*.csv")

# To make sure the files are present
print("Files found:", len(recharge_files_2019))
print(recharge_files_2019[:5]) 
```

```python
recharging_2019_list = []

for file in recharge_files_2019:
    
    # Extract country name from filename 
    country = os.path.basename(file).split(".")[0]
    
    # Read the CSV file
    df_2019 = pd.read_csv(file)
    
    # Rename the first column to 'Year'
    df_2019 = df_2019.rename(columns={df_2019.columns[0]: "Year"})
    
    # Rename the second column (which is the country name) to 'Recharging Points'
    df_2019 = df_2019.rename(columns={df_2019.columns[1]: "Recharging Points"})
    
    # Clean up Year column and keep only valid numeric years
    df_2019 = df_2019[pd.to_numeric(df_2019["Year"], errors="coerce").notnull()]
    df_2019["Year"] = df_2019["Year"].astype(int)
    
    # Add a Country column
    df_2019["Country"] = country.replace("(Target)", "").strip()
    
    # Reorder columns
    df_2019 = df_2019[["Country", "Year", "Recharging Points"]]
    
    # Append to list
    recharging_2019_list.append(df_2019)

# Combine all countries into one DataFrame
recharge_files_2019 = pd.concat(recharging_2019_list, ignore_index=True)

print("Target data combined. Shape:", recharge_files_2019.shape)
display(recharge_files_2019.head())
```

```python
# Path to all country-level Excel files
recharge_files_2025 = glob.glob(r"C:\Users\Disha\TU Braunschweig\Thesis\Data\Parameters\Recharging points (2020-2025)\*.csv")

# To make sure the files are present
print("Files found:", len(recharge_files_2025))
print(recharge_files_2025[:5]) 
```

```python
recharging_2025_list = []

for file in recharge_files_2025:
    
    # Extract country name from filename 
    country = os.path.basename(file).split(".")[0]
    
    # Read the CSV file
    df_2025 = pd.read_csv(file)
    
    # Rename first column to 'Period'
    df_2025 = df_2025.rename(columns={df_2025.columns[0]: "Period"})
    
    # Extract numeric year from 'Period' (e.g. '2020 Q1' -> 2020)
    df_2025["Year"] = df_2025["Period"].str.extract(r"(\d{4})").astype(int)
    
    # Group by year to sum all quarters (sum AC and DC, individually)
    df_2025_yearly = df_2025.groupby("Year", as_index=False)[["AC", "DC"]].sum()
    
    # Add total recharging points (AC + DC)
    df_2025_yearly["Recharging Points"] = df_2025_yearly["AC"] + df_2025_yearly["DC"]
    
    # Add country name
    df_2025_yearly["Country"] = country.replace("(Target)", "").strip()
    
    # Reorder columns
    df_2025_yearly = df_2025_yearly[["Country", "Year", "Recharging Points"]]
    
    # Append to list
    recharging_2025_list.append(df_2025_yearly)

# Combine all countries into one DataFrame
recharge_files_2025 = pd.concat(recharging_2025_list, ignore_index=True)

print("Target data combined. Shape:", recharge_files_2025.shape)
display(recharge_files_2025.head())
```

```python
# Combine both dataframe (recharge_files_2019 and recharge_files_2025)

# Make sure both dataframes exist
print("Old data:", recharge_files_2019.shape)
print("New data:", recharge_files_2025.shape)

# Ensure same column names and order
recharge_files_2019 = recharge_files_2019[["Country", "Year", "Recharging Points"]]
recharge_files_2025 = recharge_files_2025[["Country", "Year", "Recharging Points"]]

# Concatenate vertically
recharging_points = pd.concat(
    [recharge_files_2019, recharge_files_2025], 
    ignore_index=True
)

# Keep only rows where Year <= 2024 
recharging_points = recharging_points[recharging_points["Year"] <= 2024]

# Sort by Country and Year
recharging_points = recharging_points.sort_values(
    by=["Country", "Year"]
).reset_index(drop=True)

# Check the result
print("Combined dataset shape:", recharging_points.shape)
display(recharging_points.head(15))
```

```python
recharging_points['Country'].unique()
```

```python
# Keep only those countries those are part of EU27
countries_to_remove = [
    'Iceland', 'Norway', 'Liechtenstein',
    'Switzerland', 'Turkey', 'United Kingdom'
]

recharging_points = recharging_points[~recharging_points['Country'].isin(countries_to_remove)]

recharging_points['Country'].unique()
```

**Step 7: Merge these datasets with the target**

```python
# Ensure 'Year' is numeric in all datasets
targets_marketShare["Year"] = targets_marketShare["Year"].astype(int)
targets_newReg["Year"] = targets_newReg["Year"].astype(int)
gdp["Year"] = gdp["Year"].astype(int)
cpi["Year"] = cpi["Year"].astype(int)
eg["Year"] = eg["Year"].astype(int)
recharging_points["Year"] = recharging_points["Year"].astype(int)

# Ensure 'Country' is string in all datasets
targets_marketShare["Country"] = targets_marketShare["Country"].astype(str)
targets_newReg["Country"] = targets_newReg["Country"].astype(str)
gdp["Country"] = gdp["Country"].astype(str)
cpi["Country"] = cpi["Country"].astype(str)
eg["Country"] = eg["Country"].astype(str)
recharging_points["Country"] = recharging_points["Country"].astype(str)
```

```python
# Filter target dataset to years 2011-2024
targets_filtered_marketShare = targets_marketShare[(targets_marketShare["Year"] >= 2011) & (targets_marketShare["Year"] <= 2024)]

# Merge with AF New registrations (M1)
merged_w_newReg = pd.merge(targets_filtered_marketShare, targets_newReg, on=["Country", "Year"], how="left")

# Merge with GDP
merged_w_gdp = pd.merge(merged_w_newReg, gdp, on=["Country", "Year"], how="left")

# Merge with CPI
merged_w_cpi = pd.merge(merged_w_gdp, cpi, on=["Country", "Year"], how="left")

# Merge with Employment Growth
merged_eg = pd.merge(merged_w_cpi, eg, on=["Country", "Year"], how="left")

# Merge with Recharging Points
merged_df = pd.merge(merged_eg, recharging_points, on=["Country", "Year"], how="left")

# Result
print("Shape of merged dataset:", merged_df.shape)
merged_df.head()
```

```python
# Rename BEV columns
merged_df.rename(columns={
    'BEV_x': 'BEV Percentage (Total Number Of Registrations)',
    'BEV_y': 'BEV (New Registrations)'
}, inplace=True)

merged_df.head()
```
