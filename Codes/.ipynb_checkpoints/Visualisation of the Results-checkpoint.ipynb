```python
# Helper function to add metadata columns
def add_metadata(df, trial, model, cluster_type):
    df = df.copy()
    df["Trial"] = trial
    df["Model"] = model
    df["ClusterType"] = cluster_type
    return df

# Combine all datasets
all_results = pd.concat([
    
    # TRIAL 1 
    add_metadata(T1_AutoML_followers, "Trial 1", "AutoGluon", "Followers"),
    add_metadata(T1_AutoML_leaders, "Trial 1", "AutoGluon", "Leaders"),
    add_metadata(T1_lr_followers, "Trial 1", "Linear Regression", "Followers"),
    add_metadata(T1_lr_leaders, "Trial 1", "Linear Regression", "Leaders"),
    add_metadata(T1_svr_followers, "Trial 1", "SVR", "Followers"),
    add_metadata(T1_svr_leaders, "Trial 1", "SVR", "Leaders"),
    add_metadata(T1_ARIMAX_followers, "Trial 1", "ARIMAX", "Followers"),
    add_metadata(T1_ARIMAX_leaders, "Trial 1", "ARIMAX", "Leaders"),
    add_metadata(T1_rf_followers, "Trial 1", "RF", "Followers"),
    add_metadata(T1_rf_leaders, "Trial 1", "RF", "Leaders"),
    add_metadata(T1_xgb_followers, "Trial 1", "XGB", "Followers"),
    add_metadata(T1_xgb_leaders, "Trial 1", "XGB", "Leaders"),
    add_metadata(T1_lstm_followers, "Trial 1", "LSTM", "Followers"),
    add_metadata(T1_lstm_leaders, "Trial 1", "LSTM", "Leaders"),
    add_metadata(T1_proposed_ElasticNet_followers, "Trial 1", "Proposed (ElasticNet)", "Followers"),
    add_metadata(T1_proposed_ElasticNet_leaders, "Trial 1", "Proposed (ElasticNet)", "Leaders"),
    add_metadata(T1_proposed_GradientBoost_followers, "Trial 1", "Proposed (GradientBoost)", "Followers"),
    add_metadata(T1_proposed_GradientBoost_leaders, "Trial 1", "Proposed (GradientBoost)", "Leaders"),
    add_metadata(T1_proposed_Weighted_followers, "Trial 1", "Proposed (Weighted)", "Followers"),
    add_metadata(T1_proposed_Weighted_leaders, "Trial 1", "Proposed (Weighted)", "Leaders"),


    # TRIAL 2 
    add_metadata(T2_AutoML_followers, "Trial 2", "AutoGluon", "Followers"),
    add_metadata(T2_AutoML_leaders, "Trial 2", "AutoGluon", "Leaders"),
    add_metadata(T2_lr_followers, "Trial 2", "Linear Regression", "Followers"),
    add_metadata(T2_lr_leaders, "Trial 2", "Linear Regression", "Leaders"),
    add_metadata(T2_svr_followers, "Trial 2", "SVR", "Followers"),
    add_metadata(T2_svr_leaders, "Trial 2", "SVR", "Leaders"),
    add_metadata(T2_ARIMAX_followers, "Trial 2", "ARIMAX", "Followers"),
    add_metadata(T2_ARIMAX_leaders, "Trial 2", "ARIMAX", "Leaders"),
    add_metadata(T2_rf_followers, "Trial 2", "RF", "Followers"),
    add_metadata(T2_rf_leaders, "Trial 2", "RF", "Leaders"),
    add_metadata(T2_xgb_followers, "Trial 2", "XGB", "Followers"),
    add_metadata(T2_xgb_leaders, "Trial 2", "XGB", "Leaders"),
    add_metadata(T2_lstm_followers, "Trial 2", "LSTM", "Followers"),
    add_metadata(T2_lstm_leaders, "Trial 2", "LSTM", "Leaders"),
    add_metadata(T2_proposed_ElasticNet_followers, "Trial 2", "Proposed (ElasticNet)", "Followers"),
    add_metadata(T2_proposed_ElasticNet_leaders, "Trial 2", "Proposed (ElasticNet)", "Leaders"),
    add_metadata(T2_proposed_GradientBoost_followers, "Trial 2", "Proposed (GradientBoost)", "Followers"),
    add_metadata(T2_proposed_GradientBoost_leaders, "Trial 2", "Proposed (GradientBoost)", "Leaders"),
    add_metadata(T2_proposed_Weighted_followers, "Trial 2", "Proposed (Weighted)", "Followers"),
    add_metadata(T2_proposed_Weighted_leaders, "Trial 2", "Proposed (Weighted)", "Leaders"),

    # TRIAL 3 
    add_metadata(T3_AutoML_followers, "Trial 3", "AutoGluon", "Followers"),
    add_metadata(T3_AutoML_leaders, "Trial 3", "AutoGluon", "Leaders"),
    add_metadata(T3_lr_followers, "Trial 3", "Linear Regression", "Followers"),
    add_metadata(T3_lr_leaders, "Trial 3", "Linear Regression", "Leaders"),
    add_metadata(T3_svr_followers, "Trial 3", "SVR", "Followers"),
    add_metadata(T3_svr_leaders, "Trial 3", "SVR", "Leaders"),
    add_metadata(T3_ARIMAX_followers, "Trial 3", "ARIMAX", "Followers"),
    add_metadata(T3_ARIMAX_leaders, "Trial 3", "ARIMAX", "Leaders"),
    add_metadata(T3_rf_followers, "Trial 3", "RF", "Followers"),
    add_metadata(T3_rf_leaders, "Trial 3", "RF", "Leaders"),
    add_metadata(T3_xgb_followers, "Trial 3", "XGB", "Followers"),
    add_metadata(T3_xgb_leaders, "Trial 3", "XGB", "Leaders"),
    add_metadata(T3_lstm_followers, "Trial 3", "LSTM", "Followers"),
    add_metadata(T3_lstm_leaders, "Trial 3", "LSTM", "Leaders"),
    add_metadata(T3_proposed_ElasticNet_followers, "Trial 3", "Proposed (ElasticNet)", "Followers"),
    add_metadata(T3_proposed_ElasticNet_leaders, "Trial 3", "Proposed (ElasticNet)", "Leaders"),
    add_metadata(T3_proposed_GradientBoost_followers, "Trial 3", "Proposed (GradientBoost)", "Followers"),
    add_metadata(T3_proposed_GradientBoost_leaders, "Trial 3", "Proposed (GradientBoost)", "Leaders"),
    add_metadata(T3_proposed_Weighted_followers, "Trial 3", "Proposed (Weighted)", "Followers"),
    add_metadata(T3_proposed_Weighted_leaders, "Trial 3", "Proposed (Weighted)", "Leaders")
], ignore_index=True)

# Verification
print(all_results.shape)
print(all_results.columns)
all_results.head()
```

## **Note:**

Portugal will be present in leaders and followers both because - 

Portugal is in **"LEADERS" for "Trial 1 and Trial 2"**

Portugal is in **"FOLLOWERS" FOR "Trial 3"**

```python
# Model we want to plot
model_name = "Linear Regression"

# Filter data for the chosen model
df_model = all_results[all_results["Model"] == model_name]

# Compute mean absolute error per Country × Trial × Cluster
mean_error = (
    df_model.groupby(["Country", "Trial", "ClusterType"])["Error_Abs"]
    .mean()
    .reset_index()
)

# Dictionary to store Leaders and Followers
clusters = {
    "Leaders": mean_error[mean_error["ClusterType"] == "Leaders"].sort_values("Country"),
    "Followers": mean_error[mean_error["ClusterType"] == "Followers"].sort_values("Country")
}

# Leaders: pink shades, Followers: blue/purple shades
palettes = {
    "Leaders": ["#F48FB1", "#EC407A", "#C2185B"],  
    "Followers": ["#7986CB", "#5C6BC0", "#3949AB"] 
}

# Y-axis labels for clarity
ylabels = {
    "Leaders": "Mean Error (BEV %)",
    "Followers": "Mean Absolute Error (BEV %)"
}

# Visualise
sns.set(style="whitegrid", font_scale=1.1)
fig, axes = plt.subplots(2, 1, figsize=(15, 10), sharey=False)

# Loop through clusters for plotting 
for ax, (cluster_name, data) in zip(axes, clusters.items()):
    sns.barplot(
        data=data,
        x="Country",
        y="Error_Abs",
        hue="Trial",
        palette=palettes[cluster_name],  
        ax=ax
    )
    ax.set_title(f"{model_name}: Mean Absolute Error by Trial ({cluster_name.upper()})", fontsize=14)
    ax.set_ylabel(ylabels[cluster_name])
    ax.set_xlabel("Country" if cluster_name == "Followers" else "")
    ax.tick_params(axis='x', rotation=45)
    ax.legend(title="Trial", loc="upper right")
    ax.set_ylim(0, max(mean_error["Error_Abs"]) * 1.1)

plt.tight_layout()
plt.show()
```

```python
# Ensure correct data types
all_results["Trial"] = all_results["Trial"].astype(str)
all_results["Country"] = all_results["Country"].str.strip()

# Define the fixed model order
model_order = [
    "Linear Regression",
    #"SVR",
    "ARIMAX",
    "RF",
    "XGB",
    "LSTM", 
    "Proposed (ElasticNet)",
    "Proposed (GradientBoost)",
    "Proposed (Weighted)"
]

# Loop over each cluster type and plot separately
for cluster in ["Followers", "Leaders"]:
    # Filter data for the current cluster
    df_plot = all_results[all_results["ClusterType"] == cluster]

    # Create the bar plot with countries as columns and model on x-axis
    g = sns.catplot(
        data=df_plot,
        x="Model",
        y="Error_Abs",
        hue="Trial",
        col="Country",
        kind="bar",
        order=model_order,
        col_wrap=3,          
        height=8,
        aspect=2.5,
        sharey=False,
        errorbar=None, 
        palette='RdPu'
    )

    ax.legend(title="Trials", fontsize=30, title_fontsize=35, loc='upper right')
    
    # Add a title for the cluster above the plot
    g.fig.suptitle(f"Cluster Type: {cluster}", y=1.02, fontsize=50)

    # Rotate x-axis labels for readability
    for ax in g.axes.flat:
        ax.tick_params(axis="x", rotation=45, labelsize=30)
        ax.legend(title="Trial")

    plt.tight_layout()
    plt.show()
```
