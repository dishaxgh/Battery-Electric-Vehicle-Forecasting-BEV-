```python
# KPI - Median Absolute Error. 
# Reason to use this - to find the "middle" error, ignoring extreme outliers
# We group by Trial, Cluster, and Model to find the typical error performance
medae_results = all_results.groupby(['Trial', 'ClusterType', 'Model'])['Error_Abs'].median().reset_index()
medae_results.columns = ['Trial', 'Cluster', 'Model', 'MedAE (%)']

pivot_medae = medae_results.pivot_table(
    index=['Cluster', 'Model'], 
    columns='Trial', 
    values='MedAE (%)'
)

# Sort in ascending order of the Results of the latest trial (Trial 3)
pivot_medae = pivot_medae.sort_values(by=['Cluster', 'Trial 3'])

# Print Results
print("Median Absolute Error (MedAE) By Trial")
print(pivot_medae)
```

```python
# KPI - Mean Absolute Error. 
# We group by Trial, Cluster, and Model to find the typical error performance
mae_results = all_results.groupby(['Trial', 'ClusterType', 'Model'])['Error_Abs'].mean().reset_index()
mae_results.columns = ['Trial', 'Cluster', 'Model', 'MAE (%)']

pivot_mae = mae_results.pivot_table(
    index=['Cluster', 'Model'], 
    columns='Trial', 
    values='MAE (%)'
)

# Sort in ascending order of the Results of the latest trial (Trial 3)
pivot_mae = pivot_mae.sort_values(by=['Cluster', 'Trial 3'])

# Print Results
print("Mean Absolute Error (MAE) By Trial")
print(pivot_mae)
```
